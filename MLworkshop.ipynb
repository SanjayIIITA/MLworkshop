```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = plt.imread("pra.jpg")
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
function = cv2.CascadeClassifier("/content/haarcascade_frontalface_alt.xml")
```

```python
all_faces = function.detectMultiScale(img) 
```

```python
all_faces
```

```python
for f in all_faces:
    x,y,w,h = f
    img = cv2.rectangle(img,(x, y), (x+w, y+h),(255,0,0), 2)
```

```python
plt.imshow(img)
```
